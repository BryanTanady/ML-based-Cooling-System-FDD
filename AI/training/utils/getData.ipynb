```python
from SerialReader import SerialReader
import time
```

We need: 

`port`: 

**Windows:** COM1, COM2, ...

**Linux:** /dev/ttyACM0, /dev/ttyAMA0, /dev/ttyS0, /dev/ttyUSB0

**macOS:** /dev/tty.*, /dev/cu.*


`baudrate`: set the same as microcontroller

`buffer`: a list used for storing data

If we are no sure with our baudrate, we can use the method `list_devices()`

```python
devices = SerialReader.list_devices()
devices
```

Thus, the baudrate now is COM3.

Note: if there is no microcontroller connected, nothing will be returned

```python
buffer = []

reader = SerialReader(port="COM5", baudrate=9600, buffer=buffer, timeout=None)
```

My Arduino test logic prints an incrementing number once per second.

```python
print(buffer)
```

```python
import pandas as pd

time.sleep(60)

parsed = []

for line in buffer:
    line = line.strip()
    
    # Only process valid XYZ lines
    if not line.startswith("x "):
        continue

    parts = line.split()   # ['x','-0.16','y','-0.51','z','8.94']

    x_val = float(parts[1])
    y_val = float(parts[3])
    z_val = float(parts[5])
    parsed.append([x_val, y_val, z_val])

df = pd.DataFrame(parsed, columns=["X", "Y", "Z"])
df.insert(0, "Seq", range(len(df)))

df.to_csv("output.csv", index=False)
```
