```python
from SerialReader import SerialReader
import time
```

We need: 

`port`: 

**Windows:** COM1, COM2, ...

**Linux:** /dev/ttyACM0, /dev/ttyAMA0, /dev/ttyS0, /dev/ttyUSB0

**macOS:** /dev/tty.*, /dev/cu.*


`baudrate`: set the same as microcontroller

`buffer`: a list used for storing data

If we are no sure with our baudrate, we can use the method `list_devices()`

```python
devices = SerialReader.list_devices()
devices
```

Thus, the baudrate now is COM3.

Note: if there is no microcontroller connected, nothing will be returned

```python
buffer = []

reader = SerialReader(port="COM3", baudrate=9600, buffer=buffer, timeout=None)
time.sleep(5)
```

**NOTE:**

The `SerialReader` works by spawning a background thread that continuously reads incoming data, while the main thread returns immediately and continues running.

Therefore, when the reader is no longer needed, make sure to call `stop()` to properly shut down the thread and **release the serial port**

```python
reader.stop()
reader = SerialReader(port="COM3", baudrate=9600, buffer=buffer, timeout=None)
reader.stop()
```
